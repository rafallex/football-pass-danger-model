```python
import os
import json
import pandas as pd
```

```python
# Point this at your local StatsBomb open-data 'data' folder
# (https://github.com/statsbomb/open-data). Override with the STATSBOMB_DATA env var.
data_path = os.environ.get("STATSBOMB_DATA", "open-data-master/data")
```

```python
with open(os.path.join(data_path, "competitions.json")) as f:
    competitions = json.load(f)

# Convert to DataFrame for easy filtering
competitions_df = pd.DataFrame(competitions)
competitions_df.head()
```
